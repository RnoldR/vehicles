```python
agent_number = 1
winners = []

for game_index in range(number_of_games):
    
    env = connect.ConnectN()
    
    # Handle the epilon values
    epsilon *= epsilon_decay
    if epsilon < epsilon_min:
        epsilon = epsilon_min
        
    step_index = 0
    while env.is_game_over == False:
        step_index+=1
        
        # Your agents turn
        if agent_number != env.current_player:
            raise RuntimeError("Wrong player")
    
        # Get current state
        board_data = env.board.reshape(1,6,7,1)
        current_player_data = env.current_player * np.ones((1,6,7,1))
        agent_player_data = agent_number * np.ones((1,6,7,1))
        current_obsv = np.concatenate([board_data, current_player_data, agent_player_data],axis=3)
    
        # Predict Q values of current state
        current_q_values = model.predict(current_obsv)[0]
        
        # Determine action
        if random.random() < epsilon:
            # Perform random action
            action = env.get_legal_random_action()
        else:
            legal_actions = env.get_legal_actions()
            action = legal_actions[np.argmax(current_q_values[legal_actions])]
            
        # Perform action
        env.step(action)
            
        # Your agents oppenent
        if env.is_game_over == False:
            action = env.get_legal_random_action()
            env.step(action)
        
        # Get new state (this is not used when the game is over)
        board_data = env.board.reshape(1,6,7,1)
        current_player_data = env.current_player * np.ones((1,6,7,1))
        agent_player_data = agent_number * np.ones((1,6,7,1))
        next_obsv = np.concatenate([board_data, current_player_data, agent_player_data],axis=3)
        
        # Predict Q values of new state (this is not used when the game is over)
        next_q_values = model.predict(next_obsv)[0]
        
        # Handle rewards
        if env.is_game_over == False:
            reward = 0
        else:
            if env.winner == agent_number:
                reward = reward_win
            elif env.winner == 0:
                reward = reward_tie
            else:
                reward = reward_lose
                
        # Calculate new Q values    
        if env.is_game_over == True:
            current_q_values[action] = reward
        else:
            q_max_next = np.max(next_q_values)
            current_q_values[action] = reward + gamma*q_max_next
            
        # Train the model
        model.train_on_batch(current_obsv, current_q_values.reshape(1,-1))
        
    winners.append(env.winner)
    
    if len(winners) % 10 == 0:
        agent_wins = np.sum(np.array(winners[-10:]) == agent_number)
        print(game_index, agent_wins, epsilon)
        
```

## Exercise 8) 
Create an evaluation function to monitor how good the AI is doing against a random generator.

```python
def evaluate(model, agent_number, number_of_games=10):
    winners = []
    for game_index in range(number_of_games):
    
        env = connect.ConnectN()

        while env.is_game_over == False:

            # Handle the opponent
            if agent_number != env.current_player:
                action = env.get_legal_random_action()
                env.step(action)
                continue

            # Get current state
            board_data = env.board.reshape(1,6,7,1)
            current_player_data = env.current_player * np.ones((1,6,7,1))
            agent_player_data = agent_number * np.ones((1,6,7,1))
            current_obsv = np.concatenate([board_data, current_player_data, agent_player_data],axis=3)

            # Predict Q values of current state
            current_q_values = model.predict(current_obsv)[0]

            # Determine action
            legal_actions = env.get_legal_actions()
            action = legal_actions[np.argmax(current_q_values[legal_actions])]

            # Perform action
            env.step(action)
            
        winners.append(env.winner)
    return np.mean(np.array(winners) == agent_number)
```

## Exercise 11)
Create a memory class or set of functions which can do the following:
- Can hold an X amount memories
- You can add state, next_state, action, reward and if the game is over to the memory
- You can take a batch of X 
    - Bonus 1: this memory will not return empty memory states
    - Bonus 2: without the change of picking one sample twice
- FiFo: It will remove the oldest states when there are too many states in memory

```python
class Memory():
    
    def __init__(self, er_size=1000, batch_size=256, state_shape=(6, 7, 3)):
        
        self.er_size = er_size
        self.batch_size = batch_size
        
        self.states = np.zeros([er_size, *state_shape])
        self.next_states = np.zeros([er_size, *state_shape])
        self.actions = np.zeros(er_size, dtype=int)
        self.rewards = np.zeros(er_size)
        self.alives = np.zeros(er_size)
        
        self.index = 0
        self.max_index = 0
        
    def add_event(self, state, next_state, action, reward, is_alive):
        
        # Add new situation
        self.states[self.index] = state
        self.next_states[self.index] = next_state
        self.actions[self.index] = action
        self.rewards[self.index] = reward
        self.alives[self.index] = is_alive
        
        # Update indices
        self.index += 1
        if self.index > self.er_size - 1:
            self.index = 0
        self.max_index = np.max([self.index, self.max_index])
        
    def take_batch(self, batch_size=None):
        # Just to make this method a bit more flexible
        if batch_size == None:
            batch_size = self.batch_size
            
        # Inidices of batch
        num_of_samples = np.min([self.max_index, self.er_size])
        batch_size = np.min([num_of_samples, batch_size])
        indices = np.random.choice(range(num_of_samples), batch_size, replace=False)
        
        return [
            self.states[indices],
            self.next_states[indices],
            self.actions[indices],
            self.rewards[indices],
            self.alives[indices]
        ]
```
